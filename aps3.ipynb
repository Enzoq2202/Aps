<p>
O primeiro bloco escolhido pelo nosso grupo foi o Mercosul, também conhecido como Mercado Comum do Sul, que é um bloco econômico sul-americano formado pelo Brasil, Argentina, Uruguai, Paraguai, além de possuir outros países associados e observadores. O motivo de sua criação foi a tentativa de aumentar o emprego e a renda, melhorar a produtividade e intensificar as relações econômicas entre os países. No qual os critérios para os países escolhidos foram proximidade geográfica e econômica entre os países que fazem parte do bloco. Assim, os benefícios do Mercosul são a diminuição das barreiras tarifárias sobre o comércio de bens. 
<p>
O segundo bloco escolhido pelo nosso grupo foi a BRICS, que é um agrupamento econômico informal composto por cinco países: Brasil, Rússia, Índia, China e África do Sul. Os países procuraram formar um grupo político e assim, converteram seus poderes econômicos em uma maior influência geopolítica. No qual,  os critérios para a escolha dos países foram proximidade econômica e conexão de países de mercados emergentes. Deste modo, os benefícios desse bloco são a alta capacidade industrial, o potencial para se tornar o maior mercado consumidor do mundo e investimentos intensivos em infraestrutura e educação.

<font size="3" color="#00DDFF"> Probabilidade de um país daquele bloco ser avançado, isto é, P(avançado|bloco). E  utilizando o  Teorema de Bayes, P(bloco|avançado) </font>

```python
mercosul = {
    'Brasil': 'Avançado',
    'Argentina': 'Atrasado',
    'Uruguai': 'Avançado',
    'Paraguai': 'Avançado'
}

brics = {
    'Brasil': 'Avançado',
    'Rússia': 'Atrasado',
    'Índia': 'Avançado',
    'China': 'Avançado',
    'África do Sul': 'Atrasado'
}

mercosul_brics = {
    'Argentina': 'Atrasado',
    'Uruguai': 'Avançado',
    'Paraguai': 'Avançado',
    'Rússia': 'Atrasado',
    'Índia': 'Avançado',
    'China': 'Avançado',
    'África do Sul': 'Atrasado'
}


p_mercosul = 4/9
p_brics = 5/9

p_avancado_mercosul = 3/4
p_atrasado_mercosul = 1/4

p_avancado_brics = 3/5
p_atrasado_brics = 2/5

p_avancado = p_avancado_mercosul * p_mercosul + p_avancado_brics * p_brics
p_atrasado = p_atrasado_mercosul * p_mercosul + p_atrasado_brics * p_brics

p_mercosul_avancado = p_avancado_mercosul * p_mercosul / p_avancado
p_mercosul_atrasado = p_atrasado_mercosul * p_mercosul / p_atrasado

p_brics_avancado = p_avancado_brics * p_brics / p_avancado
p_brics_atrasado = p_atrasado_brics * p_brics / p_atrasado

print(f'Dado que um país é avançado, a probabilidade de ele pertencer ao Mercosul é {p_avancado_mercosul * 100:.2f} %')
print(f'Dado que um país é avançado, a probabilidade de ele pertencer ao BRICS é {p_avancado_brics * 100:.2f} %')
print(f'Dado que um país é atrasado, a probabilidade de ele pertencer ao Mercosul é {p_atrasado_mercosul * 100:.2f} %')
print(f'Dado que um país é atrasado, a probabilidade de ele pertencer ao BRICS é {p_atrasado_brics * 100:.2f} %')
print(f'Probabilidade de um país ser avançado: {p_avancado * 100:.2f} %')
print(f'Probabilidade de um país ser atrasado: {p_atrasado * 100:.2f} %')
print('|-------------------------| Exercício 4 |-------------------------|')
print(f'Probabilidade de tirar um país do Mercosul e ele ser avançado: {p_mercosul_avancado * 100:.2f} %')
print(f'Probabilidade de tirar um país do Mercosul e ele ser atrasado: {p_mercosul_atrasado * 100:.2f} %')
print(f'Probabilidade de tirar um país do BRICS e ele ser avançado: {p_brics_avancado * 100:.2f} %')
print(f'Probabilidade de tirar um país do BRICS e ele ser atrasado: {p_brics_atrasado * 100:.2f} %')
```

<font size="3" color="#00DDFF"> Sistema que classifica países como participantes de algum bloco à partir da informação de eles serem avançados ou não. </font>

```python
def classifica_pais_bloco(estado_pais):
    if estado_pais == 'Avançado':
        return 'Mercosul'
    else:
        return 'BRICS'
    
acertos = 0
erros = 0


#percorrer o dicionário e checar se o país está classificado como avancado ou atrasado
for pais, estado_pais in mercosul_brics.items():
    if classifica_pais_bloco(estado_pais) == 'Mercosul':
        if pais in mercosul:
            acertos += 1
        else:
            erros += 1
    else:
        if pais in brics:
            acertos += 1
        else:
            erros += 1
#verificar o accuracy da funcao
print(f'Accuracy da função: {acertos / (acertos + erros) * 100:.2f} %')

        
```

<p>
Conclusão: o indicador que nosso grupo escolheu não é um bom  preditor para o bloco econômico, porque o indicador que escolhemos para ver se um país é avançado ou atrasado em relação ao o Consumo de Energia Renovável (% do Consumo Total de Energia Final), é bastante falho quando comparamos países que tem uma população muito grande com Índia e China e países que tem uma população pequena como Uruguai e Paraguai, o que faz com que para podermos ter uma resposta mais concreta disso, poderíamos usar como critério 
o índice Renewable Energy Country Attractiveness Index (RECAI), no qual ele cálculo o quanto de energia renovável um país gerou por ano e quanto este país investiu, assim nós poderíamos pegar o quanto de energia foi gerada por ano em cada país e dividir por sua população e se a porcentagem fosse maior que 50% poderíamos dizer que o país está avançado neste critério.
