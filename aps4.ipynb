```python
#probability of a country from merco sul to be advanced
mercosul_adv = 0
mercosul_atr = 0
for i in mercosul:
    if mercosul[i] == 'Avançado':
        mercosul_adv += 1
    else:
        mercosul_atr += 1
print('Probability of a country from Mercosul to be advanced: ', mercosul_adv/len(mercosul))

#probability of a country from brics to be advanced
brics_adv = 0
brics_atr = 0
for i in brics:
    if brics[i] == 'Avançado':
        brics_adv += 1
    else:
        brics_atr += 1
print('Probability of a country from BRICS to be advanced: ', brics_adv/len(brics))
```
