<p>
O primeiro bloco escolhido pelo nosso grupo foi o Mercosul, também conhecido como Mercado Comum do Sul, que é um bloco econômico sul-americano formado pelo Brasil, Argentina, Uruguai, Paraguai, além de possuir outros países associados e observadores.

<p>
O segundo bloco escolhido pelo nosso grupo foi a BRICS, que é um agrupamento econômico informal composto por cinco países: Brasil, Rússia, Índia, China e África do Sul.
</p>





```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv('dados_blocos_economicos.csv', encoding='latin1')
data.head(8)
```

```python
mercosul = {
    'Brasil': 'Avançado',
    'Argentina': 'Atrasado',
    'Uruguai': 'Avançado',
    'Paraguai': 'Avançado'
}

brics = {
    'Brasil': 'Avançado',
    'Rússia': 'Atrasado',
    'Índia': 'Avançado',
    'China': 'Avançado',
    'África do Sul': 'Atrasado'
}

mercosul_brics = {
    'Argentina': 'Atrasado',
    'Uruguai': 'Avançado',
    'Paraguai': 'Avançado',
    'Rússia': 'Atrasado',
    'Índia': 'Avançado',
    'China': 'Avançado',
    'África do Sul': 'Atrasado'
}
```

<br>
<font size="3" color="#fffff">1. Para cada bloco, encontre a probabilidade de um país sorteado aleatoriamente naquele bloco ser da categoria "avançado" (isto é: P(avançado|bloco)) </font>
<br>

```python

mercosul_adv = 0
mercosul_atr = 0
for i in mercosul:
    if mercosul[i] == 'Avançado':
        mercosul_adv += 1
    else:
        mercosul_atr += 1
print('A probabilidade de um país do Mercosul ser avançado é de', (mercosul_adv/len(mercosul)) * 100, '%')


brics_adv = 0
brics_atr = 0
for i in brics:
    if brics[i] == 'Avançado':
        brics_adv += 1
    else:
        brics_atr += 1
print('A probabilidade de um país do BRICS ser avançado é de', (brics_adv/len(brics))*100, '%')
```

<br>
<font size="3" color="#fffff">2. Usando bootstrapping, calcule o intervalo de confiança de 90% para P(avançado|bloco) para cada bloco. </font>
<br>

```python
#criar um novo dataframe para o mercosul
df_mercosul = pd.DataFrame(mercosul.items(), columns=['País', 'Nível de Desenvolvimento'])
df_mercosul['Country Code'] = ['BRA', 'ARG', 'URY', 'PRY']
df_mercosul
```

```python
#criar um novo dataframe para o brics
df_brics = pd.DataFrame(brics.items(), columns=['País', 'Nível de Desenvolvimento'])
df_brics['Country Code'] = ['BRA', 'RUS', 'IND', 'CHN', 'ZAF']
df_brics
```

```python
countries_mercosul = ['BRA', 'ARG', 'URY', 'PRY']
countries_brics = ['BRA', 'RUS', 'IND', 'CHN', 'ZAF']

df = data[data['Country Code'].isin(countries_mercosul) | data['Country Code'].isin(countries_brics)]
df['Blocos Econômicos'] = np.where(df['Country Code'].isin(countries_mercosul), 'Mercosul', 'BRICS')
df_blocos = pd.concat([df_mercosul, df_brics])
df = pd.merge(df, df_blocos, on='Country Code')
df = df.drop_duplicates(subset=['Country Code', 'Blocos Econômicos'])
df
```
